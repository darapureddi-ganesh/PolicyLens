# PolicyLens — Knowledge Graph Exploration

Parses the bundled sample motor policy, builds the NetworkX knowledge graph, and inspects its structure directly — no UI, no LLM.

```python
import sys
sys.path.insert(0, "../src")

from parser import parse_policy
from entity_extractor import extract
from graph_builder import build_graph
from query_engine import answer_query

policy = parse_policy("../data/sample_motor_policy.pdf")
extracted = extract(policy)
graph = build_graph(extracted)
print(f"Clauses: {len(policy.clauses)}")
print(f"Graph: {graph.number_of_nodes()} nodes, {graph.number_of_edges()} edges")
```

```python
for u, v, data in graph.edges(data=True):
    print(u, "--", data["relation"], "-->", v)
```

```python
for q in ["Is theft covered?", "What is excluded?", "What is the deductible?", "Is engine flood damage covered?"]:
    result = answer_query(graph, q)
    print("Q:", q)
    print("A:", result.answer)
    print("Path:", result.graph_path)
    print()
```
